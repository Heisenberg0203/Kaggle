```python
df=np.log(data.ConvertedSalary+1)
df.hist(figsize=(10,5))
print(df.skew())
```

```python
##finding outliers
mean,std = np.mean(data.ConvertedSalary),np.std(data.ConvertedSalary)
lower,upper = mean-(std*3),mean+(std*3)
lower,upper,mean,std,np.min(data.ConvertedSalary),np.max(data.ConvertedSalary)
```

```python
outliers_removed = [x for x in data['ConvertedSalary'] if x > lower and x < upper]
len(outliers_removed)
```

```python
data['ConvertedSalary'].plot(kind='box',by='ConvertedSalary')
```

```python
data.ConvertedSalary.quantile([0.95])
```

### 95% Salaries are under 2,15,000 $

```python
df=data.Hobby.value_counts()
df.plot(kind='bar')
```

```python
data['ConvertedSalaryLog']= np.log(data['ConvertedSalary']+1)
```

```python
data[['Hobby','ConvertedSalaryLog']].boxplot(by='Hobby',figsize=(10,10))
```

### For both the groups median salary remains same,but there are relatively more number of people coding as Hobby

```python
data.OpenSource.value_counts().plot(kind='bar')
```

```python
data[['OpenSource','ConvertedSalaryLog']].boxplot(by='OpenSource',figsize=(10,8))
```

### Again there is not much differnce, interquantile range for both is almost same, being more people  are coding as not opensource

```python
df=pd.crosstab(data.OpenSource,data.Hobby,normalize=True,margins=True)
df
```

### Summary
1. 81% code as hobby
2. 43.5% code as openSource
3. 39% code as both hobby and opensource
4. only 4% people are opensource whose coding is not hobby

```python
df.plot(kind='bar',figsize=(20,8))
```

### Whats the most common country's of Users? 

```python
country=data.Country.value_counts()[0:15]
country.plot(kind='bar',figsize=(15,8))
```

```python
df = data[data.Country.isin(country.index)]
df[['Country','ConvertedSalaryLog']].boxplot(by='Country',figsize=(20,8),fontsize=12)
```

### Some meaningful insights
1. Us not only has most number of coders but also they have high salaries
2. India being 2nd highest has comparatively low average salary
3. Among high avg Salaries are Us,Australia,UK,Sweden

```python
df=pd.crosstab(data.Country,data.OpenSource,margins=True)
df.sort_values(by='All',ascending=False)[1:16][['Yes','No']].plot(kind='bar',figsize=(20,8))
```
